```python
def clean(doc):
    doc = re.sub("[^a-zA-Z]"," ", doc)
    words = doc.lower().split()
    words=" ".join(words)
    clean_wrd=[w for w in nlp(words) if w.is_stop!=True and w.is_punct!=True]
    word_lemma=[w.lemma_ for w in clean_wrd]
    word_lemma=" ".join(word_lemma)
    return word_lemma
    
```

```python
def word_freq(word_text,n):
    wrd_freq=Counter(word_text.split(" "))
    common_words = wrd_freq.most_common(n)
    return common_words
    
```

```python
word_freq(clean(text),5)
```
